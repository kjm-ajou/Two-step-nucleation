```python
# 1. 필요한 라이브러리 불러오기
from mp_api.client import MPRester
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# (선택 사항) Matplotlib에서 한글 폰트 설정
# font_path = 'C:/Windows/Fonts/malgun.ttf'  # 예: Windows의 맑은 고딕
# font_name = fm.FontProperties(fname=font_path).get_name()
# plt.rc('font', family=font_name)


# 2. Materials Project API에 연결
# 사용자의 API 키를 사용합니다.
api_key = "mDG44qd5yl3qzK45P7N1CsByizJjwa1n"

# --- 수정된 부분 1: FCC와 BCC 공간군 번호 목록 정의 ---
# 이전에 식별한 모든 관련 공간군 번호를 리스트로 만듭니다.
fcc_spacegroup_numbers = [202, 203, 204, 209, 210, 216, 225, 227, 228]
bcc_spacegroup_numbers = [197, 199, 204, 211, 214, 217, 220, 229, 230]
# 검색을 위해 두 리스트를 합칩니다. (중복 번호는 set을 이용해 자동 제거)
all_search_numbers = list(set(fcc_spacegroup_numbers + bcc_spacegroup_numbers))


with MPRester(api_key) as mpr:
    # --- 수정된 부분 2: 검색 필드에서 안정성 지표 제거 ---
    fields_to_get = [
        "material_id",      # 재료 ID
        "formula_pretty",   # 화학식
        "symmetry",         # 대칭성 정보 (공간군 포함)
    ]

    # --- 수정된 부분 3: 확장된 공간군 번호 리스트로 검색 ---
    docs = mpr.materials.summary.search(
        num_elements=1,
        is_metal=True,
        spacegroup_number=all_search_numbers, # 확장된 번호 리스트 사용
        fields=fields_to_get
    )

# 4. 결과 처리 및 Pandas DataFrame으로 변환
data = []
if docs:
    for doc in docs:
        # --- 수정된 부분 4: 구조 타입을 정확하게 분류 ---
        structure_type = "Unknown" # 기본값 설정
        if doc.symmetry.number in fcc_spacegroup_numbers:
            structure_type = "FCC"
        elif doc.symmetry.number in bcc_spacegroup_numbers:
            structure_type = "BCC"

        data.append({
            "원소": doc.formula_pretty,
            "MP ID": doc.material_id,
            "구조": structure_type,
            "공간군 번호": doc.symmetry.number
        })

    df = pd.DataFrame(data)
    if not df.empty:
        print("--- 단일 원소 FCC/BCC 금속 검색 결과 (전체) ---")
        print(df.to_string())
        # --- 수정된 부분 5: '원소'와 '구조'를 기준으로 중복 제거 ---
        # 이렇게 해야 한 원소가 FCC와 BCC를 모두 가질 때 둘 다 유지됩니다.
        df_unique = df.drop_duplicates(subset=['원소', '구조']).copy()
    else:
        print("검색 조건에 맞는 재료를 찾을 수 없습니다.")
        df_unique = pd.DataFrame()

else:
    print("검색 조건에 맞는 재료를 찾을 수 없습니다.")
    df_unique = pd.DataFrame()



    # 찾은 원소 목록 출력
    fcc_elements = df_unique[df_unique['구조'] == 'FCC']['원소'].tolist()
    bcc_elements = df_unique[df_unique['구조'] == 'BCC']['원소'].tolist()

    print("\n--- 발견된 원소 목록 (원소/구조 기준 중복 제거) ---")
    print(f"FCC ({len(fcc_elements)}개): {', '.join(sorted(fcc_elements))}")
    print(f"BCC ({len(bcc_elements)}개): {', '.join(sorted(bcc_elements))}")
```
